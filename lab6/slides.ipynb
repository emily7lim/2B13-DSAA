```python
import networkx as nx
import numpy as np
import matplotlib as plt
```

```python
# edge refers to connections / lines in the graph
edgeList= [ ['Udderslake','Rabbington',80],
['Rabbington', 'Beddleford',85],
['Beddleford','Wickbridge',173],
['Wickbridge','Dillchester',502],
['Dillchester','Reeshburg',84],
['Reeshburg','Udderslake',205],
['Wickbridge','Rydham',30],
['Rydham','Clamswood',183],
['Clamswood','Dillchester',167],
['Clamswood','Sheffelbrook',103],
['Sheffelbrook','Beddleford',217],
['Sheffelbrook','Bendon',186]]

g = nx.Graph() #creating an empty graph
#populating the graph with edges / lines
for edge in edgeList:
    g.add_edge(edge[0], edge[1], weight = edge[2]) #start, destination, weight
    
#visualising
nx.draw(g, pos= nx.spring_layout(g), with_labels=True )
```

```python
### DIJKSTRA'S ALGORTIHM BEHIND THE SCENES ####

# generates a list of nodes that make up the path
shortestPath= nx.shortest_path(g, 'Udderslake','Rydham', weight = 'weight')
print(shortestPath)

# calculates shortest path length (number)
dist= nx.shortest_path_length(g, 'Udderslake','Rydham', weight = 'weight')
print('Shorted path is:', shortestPath, '\nWithtotal distance of',dist,'km' )
```

```python
nx.all_pairs_dijkstra_path(g, weight = 'weight')
```

```python
# Permutates all connected cities in the graph and produces shortest path for each permutation
for paths in nx.all_pairs_dijkstra_path(g, weight = 'weight'):
    print(paths)
```

```python
# lab task 1
def displayAllShortestsPaths(g, start):
    #generator object ( to be looped through )
    all_paths = nx.all_pairs_dijkstra_path(g, weight = 'weight')
    for paths in all_paths:
        ### LINEAR SEARCH TO LOOK FOR THE start ####
        if paths[0] == start:
            for destination, path in paths[1].items():
                print('From: {} To: {} is {} km'.format(start, destination, nx.shortest_path_length(g, start,destination, weight = 'weight')))
                print(path)
displayAllShortestsPaths(g,'Rabbington')
```

```python
# lab task 3
import networkx as nx
import matplotlib.pyplot as plt

edgeList= [
    # Nodes that point to A
    ['D','A'],
    # Nodes that point to B
    ['C','B'],
    ['D','B'],
    ['E','B'],
    ['F','B'],
    ['G','B'],
    ['H','B'],
    ['I','B'],
    # Nodes that point to C
    ['B','C'],
    # Nodes that point to D
    ['E','D'],
    # Nodes that point to E
    ['F','E'],
    ['G','E'],
    ['H','E'],
    ['I','E'],
    ['J','E'],
    ['K','E'],
    # Nodes that point to F
    ['E','F'],
    # No nodes that point to G,H,I,J,K
]
    
g = nx.DiGraph() # a directed graph
for edge in edgeList:
    g.add_edge(edge[0], edge[1])

pageranks= nx.pagerank(g)
print(pageranks)

sizes = []
for key,val in pageranks.items():
    sizes.append( round(val * 100000) )
    
xspacing = 2
yspacing = 1
positions = {
    'A': (0*xspacing, 5*yspacing),
    'B': (2*xspacing, 5*yspacing),
    'C': (4*xspacing, 6*yspacing),
    'D': (0*xspacing, 3*yspacing),
    'E': (3*xspacing, 2*yspacing),
    'F': (4*xspacing, 3*yspacing),
    'G': (1*xspacing, 1*yspacing),
    'H': (2*xspacing, 0*yspacing),
    'I': (3*xspacing, 0*yspacing),
    'J': (4*xspacing, 0*yspacing),
    'K': (5*xspacing, 1*yspacing),
}

#space out the nodes
plt.box(True)
plt.tight_layout(rect = (0,0,2,2))

#add spaces all around to avoid cropping
plt.xlim(-1,11)
plt.ylim(-1,8)

# nx.draw(g, pos= nx.spring_layout(g) , with_labels=True, node_size= sizes )
nx.draw(g, 
        #nodes
        pos= positions , 
        with_labels=True, 
        node_size= sizes, 
        node_color = '#E0E0E0', 
        linewidths = 3, #thickness of the node border
        font_size = 24, 
        font_color = 'red', 
        #edges
        style = 'solid', 
        width = 1,
        edge_color = 'black', 
        )

ax = plt.gca() #get the axes of the graph

#adds border around the nodes
ax.collections[0].set_edgecolor('#000000')

#adding thr labels for importance
for key, val in pageranks.items():
    #float values but need to be formatted as a str
    #importance generated as a string
    importance = f'{val * 100:.1f}' #1dp number between 0-100
    ax.text(
        positions[key][0]+0.2, #x coord
        positions[key][1]-0.3, #y coord
        importance, #text
        fontsize = 20,
        bbox = {'facecolor': 'white', 'alpha': 1, 'pad': 2}
    )
```
